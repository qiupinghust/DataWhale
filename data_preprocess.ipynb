```python
# 检查缺失值, 过滤
tmp = df.isnull().sum()
null_sum = tmp[ tmp > 0]
drop_col = list(null_sum[null_sum > 1000].index)
df.drop(drop_col, axis = 1,inplace = True)
```

```python
df = df[df.isnull().sum(axis=1)<30]
```

```python
#缺失值填充，数值型采用均值，object后续处理
for k in df.columns:
    if df[k].dtype != "object":
        df[k].fillna(df[k].mean(), inplace=True)
    else:
        print(k)
```
